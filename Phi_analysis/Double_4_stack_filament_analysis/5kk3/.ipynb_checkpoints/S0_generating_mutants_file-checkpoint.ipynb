# Making mutant file for FoldX to run on double filament PDBs

Going to run it on 1 mutation in all 4 * 2 = 8 chains at a time
Overall that's 19aa(mutations) x 42aa(Ab length) = 798 possible mutations

Instructions/format: https://foldxsuite.crg.eu/parameter/mutant-file

```python
import pandas as pd
import numpy as np
import itertools
```

```python
path = '/lustre/scratch126/gengen/projects/amyloid_beta_epistasis/Phi_analysis/Double_4_stack_filament_analysis/5kk3/'
```

# here the structure information is only avail for aa residues 11-42 --> will only generate appropriate mutations

```python
# here we have 8 chains

# this information is from me looking at the PyMol-generated 5kk3_4_stack_double.pdb file
chains = ['A','B','C','D','J','K','L','M']
```

```python
# generating a list of all possible mutations

AB_WT = 'DAEFRHDSGYEVHHQKLVFFAEDVGSNKGAIIGLMVGGVVIA'

all_aa = ['G','A','V','L','M','I','F','Y','W','K','R','D','E','S','T','C','N','Q','H','P']

mutations = []

pos_available_5kk3 = [i for i in range(10,42)]

for pos in pos_available_5kk3:
    
    curr_Ab_wt_aa = AB_WT[pos]
    curr_prefix = curr_Ab_wt_aa + str(pos+1)
    
    for aa in all_aa:
        if curr_Ab_wt_aa != aa:
            mutations.append(curr_prefix + aa)
```

```python
# should be 32 (avail pos in structure) x 19
len(mutations)
```

```python
32*19
```

```python
# for mutation file in FoldX generate mutations in all chains

mutations4foldX_str = []

for mutation in mutations:
    print(mutation)
    curr_str = mutation[0] + chains[0] + mutation[1:] + ',' + mutation[0] + chains[1] + mutation[1:] + ',' + mutation[0] + chains[2] + mutation[1:] + ',' + mutation[0] + chains[3] + mutation[1:] + ',' + mutation[0] + chains[4] + mutation[1:] + ',' + mutation[0] + chains[5] + mutation[1:] + ',' + mutation[0] + chains[6] + mutation[1:] + ',' + mutation[0] + chains[7] + mutation[1:] + ';'
    mutations4foldX_str.append(curr_str)
```

```python
mutations4foldX_str[:10]
```

```python
with open(path + 'individual_list_5kk3_4_stack_double.txt', 'a') as the_file:
    for mutation in mutations4foldX_str:
        the_file.write(mutation+'\n')
```

```python
# generate all random combinations of chains (BDFR, BFRD, RFDB etc.) -- need 10 of these combinations for 10 FoldX runs
# not using the first combination because I will launch a smaller job above with it
combinations = list(itertools.permutations(chains, 4))[1:11]
```

```python
list(combinations[0])
```

```python
# generate 10 lines in the file for each combination of chains

mutations4foldX_str_combinations = []

for i in range(len(mutations)):
    
    curr_mutation = mutations[i]
    print(curr_mutation)
    
    for j in range(len(combinations)):
        curr_chains_list = list(combinations[j])
        
        curr_str = curr_mutation[0] + curr_chains_list[0] + curr_mutation[1:] + ',' + curr_mutation[0] + curr_chains_list[1] + curr_mutation[1:] + ',' + curr_mutation[0] + curr_chains_list[2] + curr_mutation[1:] + ',' + curr_mutation[0] + curr_chains_list[3] + curr_mutation[1:] + ';'
        
        mutations4foldX_str_combinations.append(curr_str)
```
