```python
# here will look only at runs with all 3 doubles datasets
# Sigmoid, ELU and Linear fits
# 6 l2_regularisation_values in log space

weights_Nucleation = {}

for transformation in transformations: 
    print(transformation)
        
    curr_path = results_dir + '20240219_max_terms_order_1_' + transformation + '_l2_regularization_factor_' + l2_regularisation_factor_value + '_all_3_doubles_datasets_all_variants'       


    weights_Nucleation[transformation] = pd.read_csv(curr_path + '/weights/weights_Nucleation.txt',
                                sep='\t')
```

```python
# fAD mutations
fAD_mutations = ['H6R','D7N','D7H','E11K','K16Q','L17V','A21G','E22G','E22K','E22Q','D23N','L34V','A42T']
```

```python
ddEa_for_training = {}

for transformation in transformations:
    ddEa_for_training[transformation] = pd.DataFrame(weights_Nucleation[transformation].loc[:, ['id','mean_kcal/mol']])
```

```python
np.unique(ddEa_for_training['Sigmoid']['mean_kcal/mol'].isna())
```

```python
ddEa_for_training['Sigmoid']
```

```python
# assign labels - fAD_pathogenic / not_fAD_pathogenic

for transformation in transformations:
    ddEa_for_training[transformation]['mutation_effect_label'] = ['fAD_pathogenic' if mutation in fAD_mutations else 'not_fAD_pathogenic' for mutation in list(ddEa_for_training[transformation]['id'])]
    # 1 is pathogecnic
    ddEa_for_training[transformation]['mutation_effect_binary_label'] = [1 if mutation in fAD_mutations else 0 for mutation in list(ddEa_for_training[transformation]['id'])]
```

```python
ddEa_for_training['Sigmoid']
```

```python
# make a classifier from the ddEa table for Sigmoid fit
```

```python
logreg_models = {}
y = {}
y_score = {}

for transformation in transformations:
    print(transformation)
    
    X = np.array(ddEa_for_training[transformation]['mean_kcal/mol']).reshape(-1, 1)
    y[transformation] = np.array(ddEa_for_training[transformation]['mutation_effect_binary_label'])
    logreg_models[transformation] = LogisticRegression(solver="liblinear").fit(X, y[transformation])
    logreg_models[transformation].fit(X, y[transformation])
    
    y_score[transformation] = logreg_models[transformation].predict_proba(X)[:, 1]
    
```

```python
# table of all possible 20 aa in all possible Ab (42) positions

AB_WT = "DAEFRHDSGYEVHHQKLVFFAEDVGSNKGAIIGLMVGGVVIA"
all_aa = ["G","A","V","L","M","I","F","Y","W","K","R","D","E","S","T","C","N","Q","H", "P", "*"]

ddE_df_mean_kcalmol = {}

for l2_regularisation_factor_value in l2_regularisation_factor_values:
    for transformation in transformations:
        #print(transformation)
        ddE_df_mean_kcalmol[transformation] = pd.DataFrame(columns=[elem + '-' + str(idx+1) for idx,elem in enumerate(list(AB_WT))], 
                                                           index = list(all_aa))
```

```python
ddE_df_mean_kcalmol['ELU'].head()
```

```python
# populate the ddE_df table according to weights_Nucleation

for l2_regularisation_factor_value in l2_regularisation_factor_values:
    
    for transformation in transformations:

        for idx in list(weights_Nucleation[transformation].index):

            curr_mutation_id = weights_Nucleation[transformation].loc[idx,'id']

            # continue if not WT
            if curr_mutation_id != 'WT':
                # position (1-42) of mutation in Ab sequence
                curr_mutation_pos = curr_mutation_id[1:-1]
                # aa introduced by the mutation
                curr_new_aa = curr_mutation_id[-1]

                # inferred value of ddEa to be added
                curr_ddEa_value = weights_Nucleation[transformation].loc[idx,'mean_kcal/mol']

                # minus because we have -ddEa predicted
                ddE_df_mean_kcalmol[transformation].loc[curr_new_aa, list(ddE_df_mean_kcalmol[transformation].columns)[int(curr_mutation_pos) - 1]] = -curr_ddEa_value
```

```python
# make sure values are float
for l2_regularisation_factor_value in l2_regularisation_factor_values:
    for transformation in transformations:
        for col in list(ddE_df_mean_kcalmol[transformation].columns):
            ddE_df_mean_kcalmol[transformation][col] = ddE_df_mean_kcalmol[transformation][col].astype(float)
```

```python
cmaps_ddE = {}
masks_ddE = {}

for l2_regularisation_factor_value in l2_regularisation_factor_values:
    for transformation in transformations:
        #print(transformation)
        masks_ddE[transformation] = ddE_df_mean_kcalmol[transformation].isnull()

        cmaps_ddE[transformation] = matplotlib.colormaps.get('RdBu_r')
        cmaps_ddE[transformation].set_bad("lightgrey")
```

```python
# building a ROC curve for ddEa as predictor of pathogenicity

from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_curve, roc_auc_score
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
dataset = datasets.load_wine()
X = dataset.data
y = dataset.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
clf_tree = DecisionTreeClassifier();
clf_reg = LogisticRegression();

clf_tree.fit(X_train, y_train); 
clf_reg.fit(X_train, y_train)
```

```python
y_score1 = clf_tree.predict_proba(X_test)[:,1]
y_score2 = clf_reg.predict_proba(X_test)[:,1]
```

```python
false_positive_rate1, true_positive_rate1, threshold1 = roc_curve(y_test, y_score1)
false_positive_rate2, true_positive_rate2, threshold2 = roc_curve(y_test, y_score2)
```
