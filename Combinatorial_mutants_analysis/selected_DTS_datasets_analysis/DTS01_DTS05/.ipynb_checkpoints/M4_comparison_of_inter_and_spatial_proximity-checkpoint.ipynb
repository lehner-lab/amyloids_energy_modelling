```python
structures = ['2beg','2mxu','2nao','5kk3','5oqv','7q4b','7q4m',
              '8ezd','8eze'
             ]
```

```python
chains = {}

chains['2beg'] = 'B'
chains['2mxu'] = 'D'
chains['2nao'] = 'C'
chains['5kk3'] = 'D'
chains['5oqv'] = 'F'
chains['7q4b'] = 'E'
chains['7q4m'] = 'G'
chains['8ezd'] = 'E'
chains['8eze'] = 'E'
```

```python
# read in tables with distances
results_dir = '/lustre/scratch126/gengen/projects/amyloid_beta_epistasis/amyloids_energy_modelling/DMS2structure_code_and_results/single_filament_monomer/'

physical_distances = {}

for structure in structures:
    print(structure)
    
    physical_distances[structure] = pd.read_csv(results_dir + structure + '/processed_data/PDB_contactmap_' + structure + '_monomer_' + chains[structure] + '.txt',
                                         sep=' ')
    
    physical_distances[structure] = physical_distances[structure].replace([np.inf, -np.inf], np.nan)
    physical_distances[structure] = physical_distances[structure].dropna(subset=["scHAmin"], how="all")
    
```

```python
# positions for which structural information is available
pos_available = {}
pos_available['2beg'] = [i for i in range(16,42)] # from pos 17 to 42, here enumerated from 0
pos_available['2mxu'] = [i for i in range(10,42)] # from pos 11 to 42, here enumerated from 0
pos_available['2nao'] = [i for i in range(0,42)] # from pos 1 to 42, here enumerated from 0
pos_available['5kk3'] = [i for i in range(10,42)] # from pos 11 to 42, here enumerated from 0
pos_available['5oqv'] = [i for i in range(0,42)] # from pos 1 to 42, here enumerated from 0
pos_available['7q4b'] = [i for i in range(8,42)] # from pos 9 to 42, here enumerated from 0
pos_available['7q4m'] = [i for i in range(11,42)] # from pos 12 to 42, here enumerated from 0
pos_available['8ezd'] = [i for i in range(11,42)] # from pos 12 to 42, here enumerated from 0
pos_available['8eze'] = [i for i in range(0,42)] # from pos 1 to 42, here enumerated from 0
```

```python
# in case of 2nao the first 42 positions are from one chain, and positions 43-84 - from another
# so need to mark that in the table

for structure in structures:
    print(structure)
    
    physical_distances[structure]['Pos1_chain_id'] = ['1' if 0 < Pos1 < len(pos_available[structure])+1 else '2' for Pos1 in physical_distances[structure]['Pos1']]
    physical_distances[structure]['Pos2_chain_id'] = ['1' if 0 < Pos2 < len(pos_available[structure])+1 else '2' for Pos2 in physical_distances[structure]['Pos2']]

    physical_distances[structure]['Pos1_Pos2_chain_id'] = [Pos1_chain_id + '_' + Pos2_chain_id for Pos1_chain_id, Pos2_chain_id in zip(physical_distances[structure]['Pos1_chain_id'],
                                                                                            physical_distances[structure]['Pos2_chain_id'])]
    physical_distances[structure]['Pos1_in_Abeta'] = ['' for i in range(len(physical_distances[structure]))]                                                                                        
    physical_distances[structure]['Pos2_in_Abeta'] = ['' for i in range(len(physical_distances[structure]))]        
    
    for idx in list(physical_distances[structure].index):
        
        pos1 = physical_distances[structure].loc[idx, 'Pos1']
        
        if 0 < pos1 < len(pos_available[structure])+1:
            curr_pos1_in_Abeta = AB_WT[pos_available[structure][0] + (pos1 - 1)] + '-' + str(pos_available[structure][0] + pos1)
        else:
            curr_pos1_in_Abeta = AB_WT[pos_available[structure][0] + (pos1 - len(pos_available[structure]) - 1)] + '-' + str(pos1 - len(pos_available[structure]))
        
        physical_distances[structure].loc[idx,'Pos1_in_Abeta'] = curr_pos1_in_Abeta
        
    
        pos2 = physical_distances[structure].loc[idx, 'Pos2']

        if 0 < pos2 < len(pos_available[structure])+1:
            curr_pos2_in_Abeta = AB_WT[pos_available[structure][0] + (pos2 - 1)] + '-' + str(pos_available[structure][0] + pos2)
        else:
            curr_pos2_in_Abeta = AB_WT[pos_available[structure][0] + (pos2 - len(pos_available[structure]) - 1)] + '-' + str(pos2 - len(pos_available[structure]))
        
        physical_distances[structure].loc[idx,'Pos2_in_Abeta'] = curr_pos2_in_Abeta
        
    
```

```python
# read in the flat versions of dfs with mochi 2nd order terms

mean_interaction_scores = {}

for inter_group in groups_interactions_exhaustive:
    # read and drop nans
    mean_interaction_scores[inter_group] = pd.read_csv('./files/20240722_mean_interaction_scores_' + inter_group + '_DTS01_DTS05.csv', index_col=0).dropna()
    mean_interaction_scores[inter_group]['Pos1_Pos2'] = [idx.split('_')[0][1:] + '_' + idx.split('_')[1][1:] for idx in list(mean_interaction_scores[inter_group].index)]
    
```

```python
overlap_pos_combinations = {}
mochi_terms = {}
struct_distances = {}


for structure in structures:
    
    for inter_group in groups_interactions_exhaustive:
        
        overlap_pos_combinations[structure + '_' + inter_group] = list(set(mean_interaction_scores[inter_group]['Pos1_Pos2']) & set(physical_distances[structure]['Pos1_Pos2']))
        mochi_terms[structure + '_' + inter_group] = list(mean_interaction_scores[inter_group].set_index('Pos1_Pos2').loc[overlap_pos_combinations[structure + '_' + inter_group],'mean_ddEa'])
        struct_distances[structure + '_' + inter_group] = list(physical_distances[structure].set_index('Pos1_Pos2').loc[overlap_pos_combinations[structure + '_' + inter_group],'scHAmin'])
```

```python
mochi_terms_struct_distances_dfs = {}

dddEa_thresholds = {}

spearmanr_stats = pd.DataFrame(index = structures, columns = ['spearmanr_corr','spearmanr_pvalue'])

for structure in structures:
    
    for inter_group in ['all']: # could look at all groups (groups_interactions_exhaustive)
        
        curr_key = structure + '_' + inter_group

        f, ax = plt.subplots(figsize=(5,5))

        dddEa = mochi_terms[curr_key]
        sc_dist = struct_distances[curr_key]    
        
        # top 5% of data along dddEa
        curr_dddEa_high_cutoff = np.quantile(dddEa, 0.93)
        print(structure, curr_dddEa_high_cutoff)
        
        dddEa_thresholds[structure] = curr_dddEa_high_cutoff

        
        #N_C_labels = [N_or_C_terminus(pos_combo) for pos_combo in overlap_pos_combinations[curr_key]]
        
        # will try to correlate dddEa to 1/sc_dist
        sc_dist_inverse = [1/sc_dist[i] for i in range(len(sc_dist))]
        
        
        mochi_terms_struct_distances_dfs[curr_key] = pd.DataFrame(data = {'sc_dist': sc_dist, 'dddEa': dddEa, 
                                                                          #'N_C_categ': N_C_labels, 
                                                                          'Pos1_Pos2': overlap_pos_combinations[curr_key],
                                                                         'sc_dist_inverse': sc_dist_inverse})

        sns.scatterplot(mochi_terms_struct_distances_dfs[curr_key], x='sc_dist_inverse', y='dddEa', 
                        s=25, 
                        #hue='N_C_categ', hue_order=['C_C','N_C','N_N']
                       )

        plt.grid(color='lightgrey')
        #plt.xlim(-4,4)
        #plt.ylim(-5,2)
        
        
        plt.title(curr_key + ', spearmanr stats for dddEa ~ 1/sc_dist:' + '\n' + str(scipy.stats.spearmanr(dddEa, sc_dist_inverse)))
        
        
        spearmanr_stats.loc[structure,'spearmanr_corr'] = scipy.stats.spearmanr(dddEa, sc_dist_inverse)[0]
        spearmanr_stats.loc[structure,'spearmanr_pvalue'] = scipy.stats.spearmanr(dddEa, sc_dist_inverse)[1]
        

        plt.axvline(1/8, linestyle='--', color='grey')
        #plt.axhline(curr_dddEa_high_cutoff, linestyle='--', color='grey')

        plt.ylabel('ddEa (Mochi)')
        plt.xlabel('Structure (' + structure + ') inverse distance')

        for i, txt in enumerate(list(overlap_pos_combinations[curr_key])):
            ax.annotate(txt, (sc_dist_inverse[i], dddEa[i]), fontsize=10, color='grey')
```

```python
for structure in structures:
    
    for inter_group in ['all']: # could look at all groups (groups_interactions_exhaustive)
        
        curr_key = structure + '_' + inter_group

        f, ax = plt.subplots(figsize=(2,2))

        dddEa = mochi_terms[curr_key]
        sc_dist = struct_distances[curr_key]    
        
        # top 5% of data along dddEa
        curr_dddEa_high_cutoff = 1 #np.quantile(dddEa, 0.9)
        print(structure, curr_dddEa_high_cutoff)
        
        dddEa_thresholds[structure] = curr_dddEa_high_cutoff

        
        N_C_labels = [N_or_C_terminus(pos_combo) for pos_combo in overlap_pos_combinations[curr_key]]
        
        
        sns.scatterplot(mochi_terms_struct_distances_dfs[curr_key], x='sc_dist', y='dddEa', 
                        s=5, 
                        #hue='N_C_categ', hue_order=['C_C','N_C','N_N']
                       )

        #plt.grid(color='lightgrey')
        #plt.xlim(-4,4)
        #plt.ylim(-5,2)
        
        # will try to correlate dddEa to 1/sc_dist
        sc_dist_inverse = [1/sc_dist[i] for i in range(len(sc_dist))]
        
        curr_R = scipy.stats.spearmanr(dddEa, sc_dist_inverse)[0]
        curr_pval = scipy.stats.spearmanr(dddEa, sc_dist_inverse)[1]
        
        plt.title(curr_key + ', spearman stats for dddEa ~ 1/sc_dist:' + '\n R=' + str(np.round(curr_R,2)) + ', p=' + str(np.round(curr_pval,2)),
                 fontsize=10)
        
        
        spearmanr_stats.loc[structure,'spearmanr_corr'] = curr_R
        spearmanr_stats.loc[structure,'spearmanr_pvalue'] = curr_pval
        

        plt.axvline(8, linestyle='--', color='grey', linewidth=0.5)
        plt.axhline(curr_dddEa_high_cutoff, linestyle='--', color='grey', linewidth=0.5)

        plt.ylabel('mean abs(dddEa), Mochi')
        plt.xlabel('Structure (' + structure + ') distances')

        for i, txt in enumerate(list(overlap_pos_combinations[curr_key])):
            ax.annotate(txt, (sc_dist[i], dddEa[i]), fontsize=3, color='grey')

        plt.savefig('./figures/20240722_mochi_dddEa_vs_scHA_min_structure_' + structure + '.pdf',
           bbox_inches = 'tight')
```
